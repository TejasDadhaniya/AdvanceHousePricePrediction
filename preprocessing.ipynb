```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.pandas.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('Train.csv')
test = pd.read_csv('test.csv')
print(train.shape,test.shape)
```

```python
data = pd.concat([train,test],axis=0)
```

```python
data.shape
```

```python
print('----------------------------------------')
print('Dataset contain {} rows and {} coulums'.format(data.shape[0],data.shape[1]))
print('----------------------------------------')
```

```python
numerical_features = [x for x in data.columns if data[x].dtypes != 'O']
print('----------------------------------')
print('There are "{} numerical features"'.format(len(numerical_features)))
print('----------------------------------')

categorical_features = [x for x in data.columns if data[x].dtypes == 'O']
print('There are "{} categorical features"'.format(len(categorical_features)))
print('----------------------------------')

missing_data = [x for x in data.columns if data[x].isnull().sum()>0]
print('There are "{} features" have missing values'.format(len(missing_data)))
print('----------------------------------')

missing_num_data = [x for x in missing_data if data[x].isnull().sum()>0 and data[x].dtypes != 'O' and x not in ['SalePrice']]
print('There are "{} numerical features" have missing values'.format(len(missing_num_data)))
print('----------------------------------')

missing_cat_data = [x for x in missing_data if data[x].isnull().sum()>0 and data[x].dtypes == 'O']
print('There are "{} categorical features" have missing values'.format(len(missing_cat_data)))
print('----------------------------------')
```

```python
for x in missing_num_data:
    print('{}/2919 missing value in {} '.format(data[x].isnull().sum(),x))
```

```python
for x in missing_cat_data:
    print('{}/2919 missing value in {} '.format(data[x].isnull().sum(),x))
```

```python
yr_var = [x for x in data.columns if 'Yr' in x or 'Year' in x]
yr_var
```

# Missing value

```python
#handeling missing values in numerical features
for x in missing_num_data:
    median = data[x].median()
    data[x].fillna(median,inplace=True)
data[missing_num_data].isnull().sum()    
```

```python
num = ['LotFrontage','LotArea','1stFlrSF','GrLivArea']
for x in num:
    data[x]=np.log(data[x])
```

```python
unique_cat_val = [len(data[x].unique()) for x in categorical_features]
```

```python
data.shape
```

```python
#drop feature with more than 50% missing data
for x in missing_cat_data:
    if np.round(data[x].isnull().mean(), 4) > 0.5:
        data.drop([x],axis=1,inplace=True)
```

```python
data.shape
```

```python
missing_cat = [x for x in data.columns if data[x].dtypes == 'O' and data[x].isnull().sum()>0]
len(missing_cat)
```

```python
#handling missing values in categorical features
for x in missing_cat:
    mode = data[x].mode()[0]
    data[x].fillna(mode,inplace=True)
data[missing_cat].isnull().sum()
```

```python
X = data.drop(['SalePrice','Id','YrSold'],axis=1)
y = data['SalePrice']
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
```

```python
scaler.transform(X)
```

```python
feature_scale = [x for x in X.columns]
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(X), columns=feature_scale)],
                    axis=1)
```
